```python
def primenumber(n):
    if n<2:
        return False
    for j in range(2,n//2 + 1):
        if n%j==0:
            return False
    return True
num,num1=map(int,input().split())
pm=[]
for val in range(num,num1):
    if primenumber(val):
        pm.append(val)
print(pm)
```

```python
## remove target element from original list
def remove_target(arr,target):
    n=len(arr)
    i=0
    while i<n:
        if arr[i]==target:
            arr.pop(i)
            n -=1
        else:
            i +=1
    return arr
arr=list(map(int,input().split()))
target=int(input())
print(remove_target(arr,target))
```

```python
def sort_arr(arr):
    i=0
    j=len(arr)-1
    while i<j:
        if arr[i]==2:
            arr[i],arr[j]=arr[j],arr[i]
            j -=1
        else:
            i +=1
    return arr
arr=list(map(int,input().split()))
print(sort_arr(arr))
```

```python
units = int(input())
is_senior_citizen = input().strip().lower()

bill = 0

if units <= 100:
    bill = units * 1.5
elif units <= 200:
    bill = 100 * 1.5 + (units - 100) * 2.5
elif units <= 500:
    bill = 100 * 1.5 + 100 * 2.5 + (units - 200) * 4
else:
    bill = 100 * 1.5 + 100 * 2.5 + 300 * 4 + (units - 500) * 6

# 10% subsidy for senior citizens
if is_senior_citizen in ["yes", "true", "1"]:
    bill *= 0.9

# 5% surcharge if units exceed 800
if units > 800:
    bill *= 1.05

print(f"{bill:.2f}")
```

```python
class Person:
    def __init__(self, id, name, age, dept):
        self.id = id
        self.name = name
        self.__age = age
        self.dept = dept

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age > 0:
            self.__age = age

    def details(self):
        return f"ID: {self.id}, Name: {self.name}, Age: {self.__age}, Department: {self.dept}"


class Student(Person):
    def __init__(self, id, name, age, dept, year, marks):
        super().__init__(id, name, age, dept)
        self.year = year
        self.marks = marks

    def calculate_percentage(self):
        total_marks = len(self.marks) * 100
        obtained_marks = sum(self.marks)
        return (obtained_marks / total_marks) * 100

    def details(self):
        return (f"Student -> ID: {self.id}," f"Name: {self.name}, "f"Age: {self.get_age()}, "f"Dept: {self.dept}, "f"Year: {self.year}, "f"Percentage: {self.calculate_percentage():.2f}%")

class Faculty(Person):
    def __init__(self, id, name, age, dept, subject, salary):
        super().__init__(id, name, age, dept)
        self.subject = subject
        self.__salary = salary

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        if salary >= 0:
            self.__salary = salary

    def add_teach_subject(self, subject):
        self.subject = subject

    def salary_calculation(self, leaves):
        deduction_per_leave = 500
        return self.__salary - (leaves * deduction_per_leave)

    def details(self):
        return (f"Faculty -> ID: {self.id}, "f"Name: {self.name}, "f"Age: {self.get_age()}, "f"Dept: {self.dept}, "f"Subject: {self.subject}, "f"Salary: {self.__salary}")

class Management:
    def __init__(self, college_name, country):
        self.college_name = college_name
        self.country = country
        self.students = {}
        self.faculties = {}
        self.courses = []
        self.student_id_counter = 1001
        self.faculty_id_counter = 5001

    def add_student(self, name, age, dept, year, marks):
        sid = self.student_id_counter
        self.student_id_counter += 1
        student = Student(sid, name, age, dept, year, marks)
        self.students[sid] = student
        return sid

    def remove_student(self, sid):
        if sid in self.students:
            del self.students[sid]
            print("Student Removed Successfully")
        else:
            print("Student Not Found")

    def total_students(self, dept=None, year=None):
        count = 0
        for student in self.students.values():
            if dept and student.dept != dept:
                continue
            if year and student.year != year:
                continue
            count += 1
        return count

    def add_faculty(self, name, age, dept, subject, salary):
        fid = self.faculty_id_counter
        self.faculty_id_counter += 1
        faculty = Faculty(fid, name, age, dept, subject, salary)
        self.faculties[fid] = faculty
        return fid

    def remove_faculty(self, fid):
        if fid in self.faculties:
            del self.faculties[fid]
            print("Faculty Removed Successfully")
        else:
            print("Faculty Not Found")
            
    def total_faculty(self, dept=None, subject=None):
        count = 0
        for faculty in self.faculties.values():
            if dept and faculty.dept != dept:
                continue
            if subject and faculty.subject != subject:
                continue
            count += 1
        return count

    def add_course(self, new_course):
        self.courses.append(new_course)

    def display_students(self):
        print("\n----- STUDENTS -----")
        for student in self.students.values():
            print(student.details())

    def display_faculties(self):
        print("\n----- FACULTIES -----")
        for faculty in self.faculties.values():
            print(faculty.details())

    def display_courses(self):
        print("\n----- COURSES -----")
        for course in self.courses:
            print(course)
            
college = Management("RGMCET Engineering College","India")
s1 = college.add_student("Greeshma",22,"CSE",2,[90, 85, 95, 88, 92])
s2 = college.add_student("Harsha",23,"AI",1,[80, 75, 82, 78, 85])
f1 = college.add_faculty("Madhusudhan",40,"CSE","Python",50000)
f2 = college.add_faculty("Naveen",38,"AI","Machine Learning",60000)
college.add_course("Python Full Stack")
college.add_course("Artificial Intelligence")
college.add_course("Cyber Security")
college.display_students()
college.display_faculties()
college.display_courses()
print("\nTotal Students :", college.total_students())
print("Total Faculty :", college.total_faculty())
print("\nSalary After 3 Leaves :",college.faculties[f1].salary_calculation(3))
print("\nFaculty Salary :", college.faculties[f1].get_salary())
print("Faculty Age :", college.faculties[f1].get_age())
college.remove_student(s2)
print("\nTotal Students After Removal :",college.total_students())
```
