#### math functions
 - len(iterable) : it returns the length of iterable
 - min(iterable) : it returns the smallest value from the iterable
 - max(iterable) : it returns the largest value from the iterable
 - sum(iterable) : it returns the sum of all elements in iterable

```python
arr1=[1,2,3,4,5]
arr2=['abc','abd','aCd']
print(len(arr1),len(arr2))
print(min(arr1),min(arr2))
print(max(arr1),max(arr2))
print(sum(arr1))
```

```python
#abs(val) : it returns the absolute value
print(abs(10))
print(abs(9))
```

```python
# round(val,pos)
val=10.2567890
print(round(val,2))
```

 - enumarate(iterable) : it returns the list of tuple of index and value pairs
 - map(fun,iterable) : it applies the functionality on every element
 - filter(fun,iterable) : it filter the all iterable element,it retuns the element which is satisfy the condition
 - zip(iterable1,iterable2,.....) : 

```python
lst1=[10,20,30]
lst2=['a','b','c']
print(enumerate(lst1))
print(enumerate(lst2))
```

```python
# return the target element positon in the list
lst=list(map(int,input().split()))
target=int(input())
for ind,val in enumerate(lst):
    if val==target:
        print(ind)
        break
else:
    print(-1)
```

```python
# map()
def square(x):
    return x**2
lst=[1,2,3,4,5]
print(list(map(square,lst)))
```

```python
def square(x):
    return x**2
lst=[1,2,3,4,5]
print(list(map(lambda x:x**2,lst)))
```

```python
# return the even numbers from list
def evenorodd(x):
    if x%2==0:
        return x
lst=[12,33,22,4,221,3]
print(list(filter(evenorodd,lst)))
```

```python
def evenorodd(x):
    if x%2==0:
        return x
lst=[12,33,22,4,221,3]
print(list(map(evenorodd,lst)))
```

```python
lst=[12,33,22,4,221,3]
print(list(filter(lambda x:x%2==0,lst)))
```

```python
lst=[12,33,22,4,221,3]
print(list(map(lambda x:f"{x} - even" if x%2==0 else f"{x} - odd",lst)))
```

```python
#zip(iterable1,iterable2...): it returns the tuple packing of all index wise elements
lst1=[10,20,30,40]
lst2=['a','b','c']
lst3=[1,2,3,4,5]
print(list(zip(lst1)))
print(list(zip(lst1,lst2)))
print(list(zip(lst1,lst2,lst3)))
```

```python
t=[76,98,55]
e=[58,99,65]
h=[66,85,90]
print(list(zip(t,e,h)))
```

```python
t=[76,98,55]
e=[58,99,65]
h=[66,85,90]
res=list(zip(t,e,h))
print(list(enumerate(res)))
```

 - all(iterable) : it returns True if all are non zero
 - any(iterable): it returns the True if any one element in iterable is non zero

```python
print(all([1,2,3,4]))
print(all([4,[]]))
print(all([]))
print(all([1,2,3,4,0]))
print(all([0,0,0,0]))
print(any([1,2,3,4]))
print(any([1,2,3,4,0]))
print(any([0,0,0,]))
print(any([2,[]]))
print(any([]))
```
