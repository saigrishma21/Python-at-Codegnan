#### dict methods
- value getting from dictionary
  - get(key,default_val) - it returns specified keys value if it is exits in dictionary otherwise default_value
- Add or Update dictionary items
  - update(new_dict) - it updates dictionary items with new_dict items if keys already exists otherwise adds new item
- Get keys, values, Items from dictionary
   - Keys() - it returns list of dictionary keys
   - values() - it returns list of dictionary values
   - items() - it returns list of tuple of key-value pairs
- Delete Items from dictionary
   - pop(key) - it deletes the item and returns the value of specified key, otherwise 'keyerror'
   - popitem - it delete and return the last item from the dictionary

```python
d={"name":"greeshma",'age':60,'course':'python'}
d['loc']
print(d)
```

```python
# Get
d={"name":"greeshma",'age':60,'course':'python'}
print(d.get('name'))
print(d.get('name','not exists'))
print(d.get('loc'))
print(d.get('loc','india'))
```

```python
# add or update
d={"name":"greeshma",'age':60,'course':'python'}
d.update({'age':50,'loc':'india'})
print(d)
```

```python
#get keys ,items,values
d={"name":"greeshma",'age':60,'course':'python'}
#print(d['name'])
print(d.keys())
print(d.values())
print(d.items())
```

```python
# if empty 
d={}
print(d.keys())
print(d.values())
print(d.items())
print(type(d.keys()))
print(type(d.values()))
print(type(d.items()))
```

```python
#pop(key)
d={"name":"greeshma",'age':60,'course':'python'}
print(d.pop('age'))
print(d)
print(d.pop('age'))
```

```python
# pop(item)
d={"name":"greeshma",'age':60,'course':'python'}
print(d.popitem())
print(d)
#print(d.pop('course','python'))
print(d.popitem())
print(d)
```
