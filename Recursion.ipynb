#### Direct Recursion

```python
def factorial(x):
    if x==1 or x==0:
        return 1
    return x * factorial(x-1)
x=int(input())
print(factorial(x))
```

```python
## print 1 to n numbers
def fun(n):
    if n==1:
        print(1,end=" ")
        return
    #print(n,end=" ")
    fun(n-1)
    print(n,end=" ")
fun(10)
```

```python
def fun(n):
    if n==10:
        print(10)
        return
    fun(n+1)
    print(n)
fun(1)
```

#### Indirect Recursion

```python
# find even or odd 
def even(n):
    if n == 0:
        return True
    return odd(n - 1)
def odd(n):
    if n == 0:
        return False
    return even(n - 1)
num = int(input("Enter a number: "))
if even(num):
    print("Even")
else:
    print("Odd")
```
