#### Positinal Arguments - the arguments are assigns to the paramenters as per arguments order

```python
def sub(a,b):
    return a-b
res=sub(10,5)
print(res)
print(sub(5,10))
```

#### Keyword Arguments : the arguments are passed to function calling along with the parameters

```python
def sub(x,y):
    return x-y
res=sub(x=10,y=5)
print(res)
print(sub(y=5,x=10))
```

#### Variable Length Positional Arguments

```python
def sum(x:int,*args)->int:
    print(x)
    print(type(args),args)
print(sum(10))
print(sum(10,20,30,40))
```

```python
def sum(x:int,*args)->int:
    #print(x)
   # print(type(args),args)
    s=0
    for num in args:
        s +=num
    return s+x
print(sum(10))
print(sum(x=10,20,30,40))
```
