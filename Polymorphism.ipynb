#### Method overriding

```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def show_details(self):
        return f"My name is {self.name} and iam {self.age} years old "
#child class
class Student(Person):
    def __init__(self,rollno,name,age):
        super().__init__(name,age)  # getting the properties from parent clsss
        self.rollno=rollno
    def show_details(self):
        return f"My name is {self.name} and iam {self.age} years old {self.rollno}"
s=Student(101,"sai",22)
print(s.name,s.age)
print(s.show_details())
print(s.rollno)
```

#### Method Overloading

```python
class Math:
    def add(self,a,b):
        return a+b
    def add(self,a,b,c):
        return a+b+c
    def add(self,a,b,c,d):
        return a+b+c+d
m=Math()
print(m.add(1,2,3,4))
print(m.add(1,2,3))
print(m.add(1,2))
```

```python
#same functiinons using default,varibale length position aruguments,varible lenght of keyword argument
#default arguments
class Math:
    def add(self,a,b,c=0,d=0):
        return a+b+c+d
m=Math()
print(m.add(1,2,3,4))
print(m.add(1,2,3))
print(m.add(1,2))
```
