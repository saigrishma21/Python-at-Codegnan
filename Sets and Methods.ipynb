#### set oparation or set methods
- Adding element to set
  - add(ele) - it add the element to set, if element not exists in set, otherwise nothing
  - updaet(ne_set) - it add all new set element to existing set
- Removing Element form sets
  - remove(ele) - it remove the element from set if it exists otherwise 'keyError'
  - discard(ele) - it removes the element from the set if it exits otherwise nothing
- set operation
  - union(new_sets) - It returns the common element from the all the sets
  - intersection(new_sets) - It retuns the common elements from all sets
  - difference(new_sets) - it returns the set elements those are not exists in new_Sets
  - symmetric_difference(new_sets) - it retuns the non common elements from all the sets

```python
# add and update and remove
s={1,2,3,4,5}
s.add(10)
s.add(1)
print(s)
s.update({1,2,3,4,5,56,7,8.9})
print(s)
s.remove(1)
print(s)
s.remove(100)
print(s)
```

```python
#remove,discard and add of elements
s={1,2,3,4,5}
s.add(10)
s.add(1)
print(s)
s.update({1,2,3,4,5,56,7,8.9})
print(s)
s.remove(1)
print(s)
s.discard(5)
s.discard(100)
print(s)
print(s.pop())
print(s)
```

```python
s={1,2,3,4,5}
s2={3,4,5,6,7}
print(s.union(s2))
print(s.intersection(s2))
#print(s.intersection_update(s2))
print(s.difference(s2))
print(s.symmetric_difference(s2))
print(s2.union(s))
print(s2.intersection(s))
print(s2.difference(s))
print(s2.symmetric_difference(s))
```

```python
s1={1,2,3,4,5}
s2={3,4,5,6,7}
print(s1|s2) #union
print(s1&s2) #intersection
print(s1-s2) #difference
print(s1^s2) #symmetric difference
```
