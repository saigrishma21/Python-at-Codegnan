```python
#create bank account
class BankAccount:
    def __init__(self,account,username,amount):
        self.account=account #public
        self.username=username #public
        self._amount=amount #proteced
    def getBalance(self):
        return self._amount
    def withdraw(self,withdraw_amount):
        if self._amount>=withdraw_amount:
            self._amount -=withdraw_amount
            return f"{withdraw_amount} Withdraw successfull and current {self._amount}"
        return "Insufficient Balance"
    #private method
    def __totalUsers(self):
        return "total users : 101"
sai=BankAccount(1001,"greeshma",10000)
print(sai.getBalance())
print(sai.withdraw(6000))
#accessing protected member syntax:obj.protected_member_name
print(sai._amount)
#name mangling
#accessing private members syntax: obj._classname__private member
print(sai._BankAccount__totalUsers)
print(sai._BankAccount__totalUsers())
```
