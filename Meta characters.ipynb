```python
#re.match(pat,string)
import re
string="The 3sky in blue and ocean in blue"
res=re.match('[a-zA-Z ]+',string)
print(res)
print(res.span())
print(res.group())
```

```python
#re.search(pat,string)
import re
string="The sky is blue and ocean is blue"
res=re.search('[a-zA-Z]{4}',string)
print(res)
print(res.span())
print(res.group())
```

```python
#re.findall(pat,string)
import re
string="The 3sky in blue and ocean in blue"
res=re.findall('[a-zA-Z ]{2}',string) #space also include after A-Z
print(res)
# print(res.span())
# print(res.group())
```

```python
#re.findall(pat,string)
import re
string="The sky in blue and ocean in blue"
res=re.findall('[a-zA-Z]{3,}',string)
print(res)
# print(res.span())
# print(res.group())
```

```python
#re.findall(pat,string)
import re
string="The sky is blue and ocean is blue"
res=re.findall('[a-zA-Z]..',string)
print(res)
# print(res.span())
# print(res.group())
```

```python
#re.sub(pat,new_string,string)
import re
string="The sky is blue and ocean is blue"
res=re.sub('[abcdelu]{3,}','RED',string)
print(res)
```

```python
#re.split(pat,string)
import re
string="The sky is blue 12and oce34an is bl34ue"
res=re.split('[0-9]+',string)
print(res)
```

```python
#fullmatch
# validate mobile no
import re
mobile = input()
res = re.fullmatch('^[6-9][0-9]{9}', mobile)
print(res)
if res:
    print(f"{mobile} is valid")
else:
    print(f"{mobile} is not valid")
```

```python
# fullmatch
# validate pan no
import re
pan = input()
res = re.fullmatch('^[A-Z]{5}[0-9]{4}[A-Z]$', pan)
print(res)
if res:
    print(f"{pan} is valid")
else:
    print(f"{pan} is not valid")
```

```python
#insta username
import re
iu = input()
res = re.fullmatch('^[a-z][a-z0-9_.]{7,30}', iu)
print(res)
if res:
    print(f"{iu} is valid")
else:
    print(f"{iu} is not valid")
```

```python
#fullmatch
# validate mobile no
import re
mobile = input()
res = re.fullmatch(r'^(\+91)?\s?[6-9][0-9]{9}', mobile)
print(res)
if res:
    print(f"{mobile} is valid")
else:
    print(f"{mobile} is not valid")
```

```python
#fullmatch
# validate mobile no
import re
mail = input()
res = re.fullmatch(r'^[a-z0-9][a-z0-9.]+\@[a-z0-9]+\.[a-z]{2,3}$', mail)
print(res)
if res:
    print(f"{mail} is valid")
else:
    print(f"{mail} is not valid")
```

```python
#validate passport
import re
passport = input("Enter Passport Number: ")
pattern = re.fullmatch(r'^[A-Z][0-9]{7}$',passport)
if pattern:
    print("Valid Passport Number")
else:
    print("Invalid Passport Number")
```
