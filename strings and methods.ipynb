####string methods
- case conversion methods
- lower() - it will convert entire string to lower case
- upper() - it will convert entire string to upper case
- swapcase() - it will convert lower case letter into upper case and upper case letter into lower case
- title -it converts first character of each word to upper case and remaining into lower case 
- capitalize() - it converts first character into upper case and remaining into lower case


```python
s="Forever and Ever"
print(s.lower())
print(s.upper())
print(s.title())
print(s.capitalize())
print(s.swapcase())
```

###find substring index postion
- index(val)- it return the first occurence index of substring and otherwise 'value error'
- find(val)- if substring exits it retub the foraroccurence index of substring,otherwise -1

```python
s="Forever and EveR"
print(s.index('e'))
print(s.find("e"))
print(s.find('z'))
print(s.index('z'))
```

```python
s="Forever and Ever"
print(s.index('er'))
print(s.index('er',8)) # it start checking from 8th position
```

### white space or pattern removing
- strip()-it removes the white space to both side of string
- lstrip()-it removes the left side white space to the string
- rstrip()-it removes the right side white space to the string

```python
s="      codegnan      "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s="@@@@@@@code@@@@@gnan@@@@@@@"
print(s)
print(s.strip("@"))
print(s.lstrip("@"))
print(s.rstrip("@"))
```

####count the frequency of substring
- count(sub_string)-it return the total count of non overlapping substring

```python
s="ABCACACcccc"
print(s.count('C'))
print(s.count("CAC"))
print(s.count('c'))
#count the all c (upper+lower)
print(s.count('C') + s.count('c'))
##or
str1=s.lower()
print(str1.count('c'))
##or
str1=s.upper()
print(str1.count('C'))
```

#### string checking methods
- islower() - it return True if Entire string is in lower case
- isupper()- it return True if Entire string is in upper case
- istitle()-it return True if Entire string is in title format
-iscapitalize()-it return True if Entire string is in caitalize format
- isdigit()    -it return True if Entire string contains all digits
- isalpha() - it return True if Entire string contains alphabets
- isalnum()- it return True if Entire string contains either alphabets or digits
- startswith(sub_string) -it return True if Entire string start with substring
- endswith(sub_string)- it return True if Entire string ends with substring

```python
print("code".islower())
print("Code".islower())
print("Code".isupper())
print("COdE".isupper())
print("Code123".isdigit())
print("124".isdigit())
print("code".isalpha())
print("code22".isalpha())
print("123code".isalnum())
print("code".isalnum())
print("12".isalnum())
print("Code".istitle())
print("code".istitle())
print("Codegnan".startswith("Co"))
print("Codegnan".startswith("co"))
print("Codegnan".endswith("nan"))
print("Codegnan".endswith("Co"))
```
