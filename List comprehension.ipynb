 - list comprehension syntax
 - [expression for iterator in iterable if condition]

```python
#print list 1 to n numbers
n=int(input())
res= [x for x in range(1,n+1)]
print(res)
```

```python
n=int(input())
res= [x**0.5 for x in range(1,n+1)]
print(res)
```

```python
n=int(input())
res= [round(x**0.5,4) for x in range(1,n+1)]
print(res)
```

```python
# return the list of values
# if value is even then return the square root value
# if value is odd then return the square root value
lst=list(map(int,input().split()))
res=[x**0.5 if x%2==0 else x**2 for x in lst]
print(res)
```

```python
lst=list(map(int,input().split()))
res=[round(x**0.5,4) if x%2==0 else x**2 for x in lst]
print(res)
```

```python
# two dice combinations
res=[(i,j) for i in range(1,7) for j in range(1,7)]
print(res)
print(len(res))
```

```python
res=[(i,j) for i in range(1,7) for j in range(1,7) if i==4 or j==4]
print(res)
print(len(res))
```

```python
#return the count of if sum of dices is even
res=[(i,j) for i in range(1,7) for j in range(1,7) if (i+j) %2==0 ]
print(res)
print(len(res))
```

```python
res=sum([1 for i in range(1,7) for j in range(1,7) if (i+j) %2==0 ])
print(res)
```

```python
#return sum of dices is even or odd if even 1 else 0
res=[1 if (i+j)%2==0 else 0 for i in range(1,7) for j in range(1,7)]
print(res)
```

```python
## finding whether given number is positive even or odd or negative even or odd ## using ternary operator
n=int(input())
res=("+ve even" if n%2==0 else " +ve odd") if n>=0 else ("-ve even" if n%2==0 else "-ve odd")
print(res)
```

```python
def primenumber(n):
    for i in range(2,n//2 + 1):
        if n%i==0:
            return "not prime number"
    return "prime number"
num=int(input())
print(primenumber(num))
```

```python
def primenumber(n):
    if n<2:
        return False
    for j in range(2,n//2 + 1):
        if n%j==0:
            return False
    return True
num,num1=map(int,input().split())
pm=[]
for val in range(num,num1):
    if primenumber(val):
        pm.append(val)
print(pm)
```
