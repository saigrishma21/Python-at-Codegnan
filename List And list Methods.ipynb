#### List Methods
-items adding to list
 - append(item) - it adds element at end of list
 - extend(new_list) - it adds all new_list elements at end of the list
 - insert(pos,item) - it add the item at specified position

```python
l=[1,2,3]
print(l.append(10))
print(l)
l.extend([5,6,7])
print(l)
l.insert(2,20)
print(l)
l.insert(100,50)
print(l)
l.insert(99,55)
print(l)
```

###delete items from list
- remove(item) -- it removes the specified item from the list if it exits ,otherwise  value error 
- pop(pos) -- it removes and return the value from the list for the specified position,otherwise 'index error'. By default it will selects last position 
- clear() -- it removes the all elements from the list 
- del -- it removes the either items of list from memory

```python
l=[10,2,3,4,10,2]
l.remove(10)
print(l)
l.remove(100)
print(l)
```

```python
l=[10,2,3,4,10,2]
l.remove(10)
print(l)
val=l.pop()
print(val,l)
val2=l.pop(10)
print(val2,l)
val3=l.pop(2)
print(val3,l)
l.clear()
print(l)
del l
print(l)
```
