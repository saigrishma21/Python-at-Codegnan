```python
#output formatting
#1,using old style formatting(using comma)
#name
#age
# my name is __ and  i am a ___ years old
name= "greeshma"
age= 20
marks=[90,34,22,112,33]
print("my name is",name,"and iam a",age,"years old",marks) ##old style method
print("my name is " +name+ " and iam a " +str(age)+ " years old "+str(marks)) ## type convert
```

```python
# using modulo operator 
#- %d - integer
#- %f - float
#- %s - string
name= "greeshma"
age= 20
marks=[90,34,22,112,33]
print("my name is %s and iam a %d years old %s" %(marks,age,name))
print("my name is %s and iam a %d years old %s" %(name,age,marks))
```

```python
#using dot format method
name= "greeshma"
age= 20
marks=[90,34,22,112,33]
print("my name is {} and iam {} ayears old and marks {}".format(name,age,marks))
print("my name is {} and iam {} ayears old and marks {}".format(age,name,marks))
```

```python
#using f- format
name= "greeshma"
age= 20
marks=[90,34,22,112,33]
print(f"my name is {name} and iam {age} ayears old and marks {marks}")
print(f"my name is {age} and iam {name} ayears old and marks {marks}")
```
