```python
from abc import ABC,abstractmethod
class TwoD(ABC):
    @abstractmethod
    def area(self):
        pass
    @abstractmethod
    def perimeter(self):
        pass
    @abstractmethod
    def diagonal(self):
        pass
class Square(TwoD):
    def __init__(self,side):
        self.side=side
        print("Square:")
    def area(self):
        return f" area  {(self.side)**2}"
    def perimeter(self):
        return f"perimerter {4*(self.side)}"
    def diagonal(self):
        return f"diagonal{0.5*(self.side)}"
class Rectangle(TwoD):
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
        print("Rectangle:")
    def area(self):
        return f"area  {(self.length)*(self.breadth)}"
    def perimeter(self):
        return f"perimerter  {2*((self.length)+(self.breadth))}"
    def diagonal(self):
        return f"diagonal {0.5*(((self.length)**2)+((self.breadth)**2))}"
class Triangle(TwoD):
    def __init__(self,base,b,c,height):
        self.base=base
        self.b=b
        self.c=c
        self.height=height
        print("Triangle:")
    def area(self):
        return f" area  {0.5*((self.base)*(self.height))}"
    def perimeter(self):
        return f"perimerter  {((self.base)+(self.b)+(self.c))}"
    def diagonal(self):
        return f"in triangle diagonal does not perform"
class Circle(TwoD):
    def __init__(self,radius):
        self.radius=radius
        print("Circle:")
    def area(self):
        return f" area  {3.14*((self.radius)**2)}"
    def perimeter(self):
        return f"perimerter  perform"
    def diagonal(self):
        return f"in circle does not perform diagonal"
    def circumference(self):
        return f" circumference  {2*3.14*(self.radius)}"
    def diameter(self):
        return f" diameter  {(2*(self.radius))}"
s=Square(2)
print(s.area())
print(s.perimeter()) 
print(s.diagonal())
r=Rectangle(5,10)
print(r.area())
print(r.perimeter()) 
print(r.diagonal())
t=Triangle(5,8,3,10)
print(t.area())
print(t.perimeter()) 
print(t.diagonal())
c=Circle(5)
print(c.area())
print(c.perimeter())
print(c.diagonal())
print(c.circumference())
print(c.diameter())
```

```python
from abc import ABC,abstractmethod
class ThreeD(ABC):
    @abstractmethod
    def volume(self):
        pass
    def surface_area(self):
        pass
class Cube(ThreeD):
    def __init__(self,side):
        self.side=side
        print("Cube:")
    def volume(self):
        return f"Volume:{(self.side)*3}"
    def surface_area(self):
        return f"Surface area:{6*((self.side)**2)}"
class Cylinder(ThreeD):
    def __init__(self,radius,height):
        self.radius=radius
        self.height=height
        print("Cylinder:")
    def volume(self):
        return f"Volume:{3.14*(((self.radius)**2)*(self.height))}"
    def surface_area(self):
        return f"Surface area:{2*3.14*(self.radius)*((self.radius)+(self.height))}"
class Cone(ThreeD):
    def __init__(self,radius,height,length):
        self.radius=radius
        self.height=height
        self.length=length
        print("Cone:")
    def volume(self):
        return f"Volume:{(1/3)*(3.14*((self.radius)**2)*(self.height))}"
    def surface_area(self):
        return f"Surface area:{3.14*(self.radius)*((self.radius)+(self.length))}"
class Sphere(ThreeD):
    def __init__(self,radius):
        self.radius=radius
        print("Sphere:")
    def volume(self):
        return f"Volume:{(4/3)*(3.14*((self.radius)*3))}"
    def surface_area(self):
        return f"Surface area:{4*3.14*((self.radius)**2)}"
c=Cube(3)
print(c.volume())
print(c.surface_area())
c1=Cylinder(3,5)
print(c1.volume())
print(c1.surface_area())
c2=Cone(3,5,6)
print(c2.volume())
print(c2.surface_area())
s=Sphere(3)
print(s.volume())
print(s.surface_area())
```
