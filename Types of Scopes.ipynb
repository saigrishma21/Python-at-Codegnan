```python
x=10
def outer():
    global x
    y=20
    x=x+y
    print("Local scope:",x+y)
    def inner():
        z=30
        print("Non Local scope:",x+y+z)
    inner()
outer()
print("global scope:",x)
```

```python
x=10
def outer():
    global x
    y=20
    x=x+y
    def inner():
        nonlocal y
        z=30
        y=y+z+x
        print("Non Local scope:",x+y+z)
    inner()
    print("Local scope:",y)
outer()
print("global scope:",x)
```

```python
#pass by value
def counter(x):
    print("----------function execution start-----------")
    print(f"Before updating x= {x} value reference {id(x)}")
    print(id(x))
    x=x+1
    print(f"After operation x ={x} reference: {id(x)}")
    print("-------------Function execution completed----------")
a=10
print(f"Before function execution{a}value referrence:{id(a)}")
counter(a)
print("After function execution:",a)
```

```python
#pass by reference
def counter(x):
    x.append(5)
a=[1,2,3,4]
counter(a)
print(a)
```

```python
## item assignment
def counter(x):
    print("----------function execution start-----------")
    print(f"Before updating x= {x} value reference {id(x)}")
    print(id(x))
    x=[10,20,30] #item assignment
    print(f"After operation x ={x} reference: {id(x)}")
    print("-------------Function execution completed----------")
a=[1,2,3]
print(f"Before function execution{a}value referrence:{id(a)}")
counter(a)
print("After function execution:",a)
```
