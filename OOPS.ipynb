```python
,#Student 
class Student:
    school="ABC SCHOOL"  #class attribute
    def __init__(self,name,age,branch="CSE"):#method- initialisation method/constructor(magic method)
        self.name=name  #attributes
        #self.rollno=rollno
        self.age=age
        self.branch=branch
        print("__inti__ constructor executed")
    def details(self): #method
        d=dict()
        d['Name']=self.name
        d['Age']=self.age
        d['Branch']=self.branch
        d['School']=Student.school
        return d
        #class method 
    @classmethod
    def update_school_name(cls,new_name):
        cls.school=new_name
    #satic method defination
    @staticmethod    
    def calculate_percentage(marks:list):
        percentage=(sum(marks)/300)*100
        return round(percentage,2)

s1=Student("Sai",22,"ECE")
s2=Student("Greeshma",24,"CSE")
print(s1)
print(s1.details())
#print(s1.calculate_percentage([85,69,78]))
Student.update_school_name("XYZ School")
print("after updating school name")
print(s1.school, s2.school)
print(Student.calculate_percentage([85,5,33]))
```
