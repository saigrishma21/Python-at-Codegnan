```python
from datetime import datetime
today=datetime.now()  ## it returns the current time stamp
print(today)
print(today.date())  # by using the variable today
print(datetime.now().date())  ##import from the module
print(today.time()) ##by using the variable today
print(datetime.now().time()) #importing from the module
```

```python
date=today.date()
time=today.time()
print(date)
print(time)
print(date.day)
print(date.month)
print(date.year)
print(time.hour)
print(time.minute)
print(time.second)
```

```python
print(date.min)
print(date.max)
```

```python
print(date.strftime("%d-%m-%Y"))
print(date.strftime("%d-%m-%y"))
print(date.strftime("%d-%m-%Y")[:5])
print(date.strftime("%B %d,%Y"))
print(time.strftime("%H:%M,%p"))
print(date.replace(month=3,year=2004,day=18))
```

```python
#Timer Implementation
import time
from IPython.display import clear_output
sec_counter=min_counter=hour_counter=0
while True:
    print(f"{hour_counter} : {min_counter} : {sec_counter}")
    time.sleep(1)
    sec_counter +=1
    if sec_counter ==60:
        min_counter +=1
        sec_counter=0 
        if min_counter ==60:
            hour_counter +=1
            min_counter=0
            if hour_counter==24:
                hour_counter=0
    clear_output(wait=True)
```
