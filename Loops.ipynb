```python
#count how many vowels and consonants
n=input().lower()
v=0
c=0
s=n.replace(" ","")
for i in s:
    if i in "aeiou":
        v +=1
    else:
        c +=1
print(s)
print(v)
print(c)
    
```

```python
n=input().lower()
v=0
c=0
s=n.replace(" ","")
for i in s:
    if i.isalpha():
        if i in "aeiou":
            v +=1
        else:
            c+=1
print(v)
print(c)
```

```python
#check whetheaar the element is exists or not if exists check the index
l=list(map(int,input().split()))
target=int(input())
n=len(l)
for i in range(n):
    if target==l[i]:
        print(i)
        break
else:
    print(-1)
```

```python
l=list(map(int,input().split()))
target=int(input())
n=len(l)
for i in range(n):
    if target==l[i]:
        print(i)
        continue
else:
    print(-1)
```

#### Nested Loops

```python
for i in range(1,5):
    for j in range(1,3):
        print(i,j)
```

```python
for i in range(1,5):
    for j in range(1,3):
        print("*")
```

```python
n=5
for i in range(n):
    for j in range(n):
        print("*",end=" ")
  
```

```python
n=5
for i in range(n):
    for j in range(n):
        print("*",end=" ")
    print()
```
