```python
## solve add,sub,mul,div, using operators
a=int(input())
b=int(input())
print(a+b)
print(a-b)
print(a*b)
print(a//b)
print(a%b)
```

```python
#using if elif
bill_amount = float(input("Enter the total bill amount: "))
if bill_amount < 500:
    print("No discount")
elif 500 <= bill_amount <= 999:
    print("5% discount")
elif 1000 <= bill_amount <= 1999:
    print("10% discount")
else:
    print("20% discount")
```

```python
## converting string sentence to lower by particular word
sentence = input()
word = input()

if word.lower() in sentence.lower():
    words = sentence.split()

    for i in range(len(words)):
        if words[i].lower() == word.lower():
            words[i] = words[i].upper()

    print(" ".join(words))
else:
    print(sentence.lower())
```

```python
### list 
n = int(input())
scores = list(map(int, input().split()))

if n > 5:
    scores.sort(reverse=True)
    print("Top 3:", scores[:3])
else:
    print("Not enough scores")
```

```python
##dict
name = input()
marks = int(input())
grade = input()

student = {
    "name": name,
    "marks": marks,
    "grade": grade
}

if marks >= 70:
    student["status"] = "Pass"
    student["scholarship"]=True
    print(f"{student['name']}->{student['status']}")
    print("Keys:",list(student.keys()))
else:
    print("No update required")
```

```python
##set
python_students = set(input().split())
java_students = set(input().split())

both = python_students.intersection(java_students)
only_python = python_students.difference(java_students)

if both:
    print("Both Courses:", sorted(both))
    print("Only Python:", sorted(only_python))
else:
    print("No common students")
```

```python
#tuple
sales = tuple(map(int, input().split()))

if max(sales) >= 50000:
    print("Highest Sales:", max(sales))
    print("Lowest Sales:", min(sales))
    print("Average Sales:", round(sum(sales) / len(sales), 2))
else:
    print("Sales target not achieved")
```
