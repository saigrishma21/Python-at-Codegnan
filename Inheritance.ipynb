#### Single Inheritance
 - if child class doesnot have any constructor the parent clss constructor acts as child class constructor

```python
#only parent class implementation without child class implementation
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def show_details(self):
        return f"My name is {self.name} and iam {self.age} years old"
#child class
class Student(Person):
    pass
s=Student("sai",22)
print(s.name,s.age)
print(s.show_details())
```

```python
#only parent class implementation with child class implementation
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def show_details(self):
        return f"My name is {self.name} and iam {self.age} years old"
#child class
class Student(Person):
    def __init__(self,rollno):
        self.rollno=rollno
    
s=Student(101)
t=Person("sai",20)
print(t.name,t.age)
print(t.show_details())
print(s.rollno)
```

```python
#only parent class implementation with child class implementation
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def show_details(self):
        return f"My name is {self.name} and iam {self.age} years old "
#child class
class Student(Person):
    def __init__(self,rollno,name,age):
        self.rollno=rollno
        self.name=name
        self.age=age
    
s=Student(101,"sai",22)
print(s.name,s.age)
print(s.show_details())
print(s.rollno)
```

```python
#inherting the parent class constructor initialization child class constructor
#super() : the super() builtin function is used to get the parent clss properties initialization in child class
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def show_details(self):
        return f"My name is {self.name} and iam {self.age} years old "
#child class
class Student(Person):
    def __init__(self,rollno,name,age):
        super().__init__(name,age)  # getting the properties from parent clsss
        self.rollno=rollno
s=Student(101,"sai",22)
print(s.name,s.age)
print(s.show_details())
print(s.rollno)
```

```python
#inherting the parent class constructor initialization child class constructor
#super()
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def show_details(self):
        return f"My name is {self.name} and iam {self.age} years old "
#child class
class Student(Person):
    def __init__(self,rollno,name,age):
        super().__init__(name,age)  # getting the properties from parent clsss
        self.rollno=rollno
    def calculate_percentage(self,marks:list):
        percentage=sum(marks)/len(marks)
        return f" {self.name} percentage is {percentage:.2f}"
p=Person("Greeshma",21)
s=Student(101,"sai",22)
print(p.show_details())
print(s.show_details())
print(s.calculate_percentage([10,9,8]))
```

#### Multi level inheritance

```python
class Employee:
    def __init__(self, name):
        self.name = name
        #print("Employee Constructor")

    def work(self):
        print(f"{self.name} performs regular employee tasks")


class Manager(Employee):
    def __init__(self, name, department):
        super().__init__(name)
        self.department = department
        #print("Manager Constructor")

    def manage_team(self):
        print(f"{self.name} manages the {self.department} team")


class ProjectManager(Manager):
    def __init__(self, name, department, project):
        super().__init__(name, department)
        self.project = project
        #print("Project Manager Constructor")

    def manage_project(self):
        print(f"{self.name} is handling the project '{self.project}'")

    def display(self):
        #print("\nEmployee Details")
        #print("Name:", self.name)
        #print("Department:", self.department)
        #print("Project:", self.project)
        return f" my name is {self.name} iam from {self.department} department and handling the project is {self.project}"


pm = ProjectManager("Greeshma", "IT", "Online Shopping System")

print()

print(pm.work())           
print(pm.manage_team())      
print(pm.manage_project())   
print(pm.display())
```

```python
class Employee:
    def __init__(self,id,name,salary,role="Employee"):
        self.id=id
        self.name=name
        self.salary=salary
        self.role=role
    def employee_details(self):
        return f" Employee id: {self.id} - Name: {self.name} and iam working as {self.role} and my salary is {self.salary}"
        
class TeamLead(Employee):
    def __init__(self,id,name,salary,team_size,project_name,role="Team Lead"):
        super().__init__(id,name,salary,role)
        self.team_size=team_size
        self.project_name=project_name
    def project_info(self):
        return f" I am handling {self.project_name} with {self.team_size} employee" 

class Manager(TeamLead):
    def __init__(self,id,name,salary,team_size,project_name,no_projects,role="Manager"):
        super().__init__(id,name,salary,team_size,project_name,role)
        self.no_projects=no_projects
    def project_status(self,status):
        return f"{self.project_name} project status is {status}" 
m=Manager("emp101","greeshma",100000,8,"cms portal",10)
print(m.project_info())
print(m.employee_details())
print(m.project_status("completed"))
```

#### Multiple Inheritance

```python
#MRO method resolution operator
class Father:
    def __init__(self,fathername):
        self.fathername=fathername
    def show1(self):
        return "this is father class"
class Mother:
    def __init__(self,mothername):
        self.mothername=mothername
    def show2(self):
        return "this is mother class"
class Child(Father,Mother):
    def __init__(self,fathername,mothername,childname):
        super().__init__(fathername)
        super().__init__(mothername)
    def details(self):
        print(super().show1())
        print(super().show2())
c=Child("rajaskehar","mallika","greeshma")
print(Child.mro())
print(c.details())
```

```python
#MRO method resolution operator
class Father:
    def __init__(self,fathername):
        self.fathername=fathername
    def show1(self):
        return "this is father class"
class Mother:
    def __init__(self,mothername):
        self.mothername=mothername
    def show2(self):
        return "this is mother class"
class Child(Father,Mother):
    def __init__(self,fathername,mothername,childname):
        Father.__init__(self,fathername)
        Mother.__init__(self,mothername)
    def details(self):
        print(super().show1())
        print(super().show2())
c=Child("rajaskehar","mallika","greeshma")
print(Child.mro())
print(c.details())
```
