```python
# generator
def generator(x):
    x=x-1
    yield x
    x=x-1
    yield x
    x=x-1
    yield x
res=generator(5)
print(res)
print(next(res))
print(next(res))
print(next(res))
```

```python
#generator
# 
def generator(x):
    x=x-1
    yield x
    x=x-1
    yield x
    x=x-1
    yield x
res=generator(5)
for val in res:
    print(val)
```

```python
def print_number(n):
    for i in range(n):
        yield i 
res=print_number(11)
for val in res:
    print(val)
```

```python
def fun(n):
    if n==1:
        print(1,end=" ")
        return
    #print(n,end=" ")
    fun(n-1)
    print(n,end=" ")
fun(10)
```
